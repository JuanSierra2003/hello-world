```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def filon_quadrature(w_range, signal_fn, num_intervals, time):
    w = np.linspace(*w_range, num_intervals)
    dw = w[1] - w[0]
    integral_real = np.sum(signal_fn(w) * np.cos(w * time)) * dw
    integral_imag = np.sum(signal_fn(w) * np.sin(-w * time)) * dw
    return complex(integral_real, integral_imag)
```

```python
def signal(w, mu, sigma, alphap):
    return np.exp(-.5/(sigma**2)*(w-mu)**2 + .5j*alphap*(w-mu)**2)
```

```python
mu, sigma, alphap = 2.7, .5, 100
w_range = (mu-6*sigma, mu+6*sigma)
w = np.linspace(*w_range, 1000)
```

```python
plt.plot(w, np.abs(signal(w, mu, sigma, alphap)))
```

```python
time = np.linspace(-300, 300, 10000)
```

```python
plt.plot(time, np.real([filon_quadrature(w_range, lambda w: signal(w, mu, sigma, alphap), 1000, tt) for tt in time]))
```
