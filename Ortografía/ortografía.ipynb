```python
import google.generativeai as genai

# Configure the API key# Send to Gemini
genai.configure(api_key=GEMINI_API_KEY)
# Import the GenerativeModel class from the Google Generative AI library 

# Initialize the model (e.g., 'gemini-pro' for text)
model = genai.GenerativeModel('gemini-2.0-flash')
```

```python
def correct(text: str) -> str:
    """
    Corrects the spelling, grammar, and punctuation of the given text using the Gemini model.
    Args:
        text (str): The text to be corrected.
    Returns:
        str: The corrected text with modifications marked in brackets.
    """
    prompt = f"""
    Eres un editor de texto experto. Tu tarea es corregir la ortografía, gramática y puntuación del siguiente código Latex, manteniendo el significado original:
    {text}
    Devuelve el texto corregido, sin ningún comentario adicional.
    No alteres los comandos como equaciones o expresiones matemáticas. Déjalo tal y como es, sin etiquetas adicionales.
    """
    
    # Inicializar modelo de Gemini
    response = model.generate_content([
        {"text": prompt},
    ])
    return response.text
```

```python
def correct_file():
    """
    Reads input from 'input.txt', corrects it using the Gemini model, and writes the output to 'output.txt'.
    """
    # Leer el texto de entrada desde 'input.txt', corregirlo y escribir el resultado en 'output.txt'
    input_text = open('input.txt', 'r').read().split('\n\n')
    print(f"Iniciando la corrección del texto...{len(input_text)} párrafos encontrados.")

    corrected_text = ""
    for ii, paragraph in enumerate(input_text):
        if paragraph.strip():  # Verificar si el párrafo no está vacío
            while True:
                try:
                    # Intentar corregir el párrafo
                    corrected_paragraph = correct(paragraph)
                    break  # Si la corrección es exitosa, salir del bucle
                except Exception as e:
                    print(f"Error al corregir el párrafo {ii + 1}/{len(input_text)}: Reintentando...")
                
            # Si la corrección es exitosa, agregar el párrafo corregido al texto final
            corrected_text += corrected_paragraph + "\n\n"
        print(f"Corrección del párrafo {ii + 1}/{len(input_text)} completada.")
        
    corrected_text = corrected_text.strip()  # Eliminar el último salto de línea extra
    print("Corrección del texto completada.")
    
    # Escribir el texto corregido en 'output.txt'
    corrected_text = corrected_text

    with open('output.txt', 'w') as f:
        f.write(corrected_text)
    print("Corrección completada. El texto corregido se ha guardado en 'output.txt'.")
```

```python
correct_file()
```
