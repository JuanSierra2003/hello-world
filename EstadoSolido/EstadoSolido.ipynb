## **Teoría**
La anisotropía magnética hace refrencia a la dependencia direccional de las propiedades magnéticas de un material.
Esta puede ser tener varios orígenes como:
- Anisotropía de forma.
- Anisotropía por estrés.
- Anisotropía inducida por deformación plástica, aleaciones e irradiación.
- Anisotropía de intercabio.
- **Anisotropía de cristal**
    
## Anisotropía de cristal
### Origen
**1. Campo cristalino (fuerte interacción red-órbita)**
La presencia de átomos vecinos crea un campo eléctrico que limita el movimiento del electrón a direcciones específicas dentro del cristal. Como resultado las órbitas de los electrones tienden a elongarse en ciertas direcciones respecto al cristal (estas son las **direcciones fáciles**). 
Esta interacción es tan fuerte que cuando se aplica un $\vec{H}$ a lo largo de alguna dirección arbitraria en el cristal, impide que las órbitas, y sus momentos magnéticos orbitales asociados, giren hacia la dirección de $\vec{H}$. Esto es conocido como *orbital quenching* lo que significa que el momento orbital se considera efectivamente nulo al pensar en las propiedades magneticas del material.
$$\vec{M}=-\mu_B(\vec{L}+2\vec{S}) \to \vec{M}_{eff}=-2\mu_B\vec{S}$$
**2. Interacción espín-órbita (débil interacción)**
Esta interacción $\sim \vec{L}\cdot \vec{S}$ tiende a alinear los espines a las órbitas y lo que en suma orienta los espines a las direcciones fáciles de la red (por **1.**). 
Sin embargo esta interacción es usualmente débil, por lo que un $\vec{H}$ puede superar esta interacción y alinear los espines hacia $\vec{H}$. La energía necesaria para romper esta interacción es llamada la **enegía anisotrópica**.

### Enegía anisotrópica
Esta es expresada como una serie de potencias de las direcciones coseno de $\vec{M}$ respecto al cristal. Pero como $\vec{M}$ es impar respecto a la inversión temporal y la energía es par, entonces esta serie solo tiene términos **pares**.  
Adicionalmente, teniendo en cuenta que la energía debe mantenerse **invariante bajo las simetrías del cristal**, se obtiene que por ejemplo:
- *Uniaxial*: $E_{anis} = K'_1\cos^2(\theta)+...=K_1\sin^2(\theta)+...$
- *Cúbica*: $E_{anis} = K_1 \left( \alpha_1^2 \alpha_2^2 + \alpha_2^2 \alpha_3^2 + \alpha_3^2 \alpha_1^2 \right) + K_2 \left( \alpha_1^2 \alpha_2^2 \alpha_3^2 \right)+...$ 
con $\alpha_i$ el coseno del ángulo con el eje $i$. 

La $E_{anis}$ tiene una dependencia **decreciente con la temperatura**, llegando a anularse en el punto de Curie ($T=T_C$). 
Los valores de las constantes $K$ rigen qué ejes son fáciles (energía minima) o difíciles (energía máxima).

## **Aplicaciones**

Los imanes anisotrópicos son imanes cuyas propiedades magnéticas dependen de su dirección de magnetización. Esto significa que tienen diferentes niveles de magnetismo en distintas direcciones de magnetización. Cuando se fabrican, se alinean de tal manera que tienen una dirección de magnetización preferida. Fuera de esta dirección, no se pueden magnetizar. Una ventaja de este tipo de imán es que tiene propiedades magnéticas más fuertes que los imanes isotrópicos.


En un sistema simplicado uniaxial, el campo de coerción esta dado por
$$H_c\approx \frac{2K_1}{\mu_0M_s},$$
Luego una gran anisotropía (gran $K_1$) implica un gran $H_c$ indicando que es un imán duro y viceversa para los imanes blandos.

**1. Imanes duros**

Son aquellos que  tienen una alta coercitividad, lo que significa que una vez magnetizados retienen su magnetismo durante un largo tiempo y es difícil desmagnetizarlos. La mayoría los imanes duros son anisotrópicos. Sin embargo, esto no significa que todos los imanes anisotrópicos sean imanes duros ni que todos los imanes duros sean anisotrópicos. 

- Un ejemplo son los imanes de neodimio sinterizado, producidos de tal manera que los granos están orientados en una dirección preferida para la magnetización. Los imanes de neodimio están hechos de neodimio, hierro y boro. Para producirlos, primero se magnetiza la aleación y se vierte en un molde donde se enfría. En el siguiente paso, la aleación se muele hasta convertirla en polvo fino. El polvo se coloca en un campo magnético y se le da forma bajo alta presión. Este paso de la fabricación de neodimio sinterizado es muy importante porque determina la dirección de magnetización preferida del imán. Posteriormente, cuando los imanes terminados se colocan en una bobina magnética, el campo magnético de la bobina estará en la misma dirección que la red cristalina de los imanes de neodimio sinterizado. Es este campo magnético de la bobina el que magnetizará los imanes de neodimio sinterizado de forma permanente. 

  Este tipo de imanes se utiliza para aumentar la eficiencia en *motores eléctricos* (p. ej., en vehículos eléctricos y motores industriales); en *turbinas eólicas*; en unidades de *disco duro*; para activar los diafragmas de los *altavoces* con campos magnéticos fuertes mejorando la calidad del sonido en altavoces y auriculares; en *resonancia magnética* proporcionan campos magnéticos fuertes y estables para imágenes médicas de alta resolución, con ellos también se detecta el espín de protones y neutrones.

- Otro ejemplo de imanes anisotrópicos son los de samario y cobalto, que se utilizan en *sensores*. Funcionan transformando un campo magnético en señales eléctricas, suelen emplearse en la industria aeroespacial y automovilística.

**2. Imanes blandos**
Son aquellos que se magnetizan fácilmente pero pierden el magnetismo rápidamente al retirar el campo magnético. Como ejemplos se encuentran los siguientes:
- Ferritas anisotrópicas (NiZn, MnZn) usadas en transformadores de alta frecuencia en computadoras, cargadores de teléfonos, televisores.
- Filtros de ruido electromagnético: en inductores y componentes de filtro para eliminar el ruido electromagnético en circuitos electrónicos, protegiendo dispositivos como routers Wi-Fi, televisores y consolas de videojuegos.
- Acero al silicio con orientación de grano: usado en transformadores de potencia en subestaciones eléctricas y plantas generadoras de electricidad; reducen las pérdidas por corrientes parásitas y mejoran la eficiencia de la transmisión de energía.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define anisotropy energy surface for cubic crystal
def ecubic(theta, phi, K0, K1, H):
    return (K0 + K1 * ((np.sin(theta) * np.cos(phi))**2 * 
            ((np.sin(theta) * np.sin(phi))**2 + (np.cos(theta))**2 * (np.sin(theta) * np.sin(phi))**2 + 
             (np.cos(theta))**2 * (np.sin(theta) * np.cos(phi))**2)) + 
            (H / np.sqrt(3)) * (np.sin(theta) * np.cos(phi) + np.sin(theta) * np.sin(phi) + np.cos(theta)))

# Energy projections along x, y, and z
def fcubicx(theta, phi, K0, K1, H):
    return np.sin(theta) * np.cos(phi) * ecubic(theta, phi, K0, K1, H)

def fcubicy(theta, phi, K0, K1, H):
    return np.sin(theta) * np.sin(phi) * ecubic(theta, phi, K0, K1, H)

def fcubicz(theta, phi, K0, K1, H):
    return np.cos(theta) * ecubic(theta, phi, K0, K1, H)

#SC

# Set anisotropy constants
K0 = 1
K1 = 0

# Create grid for x and y (in spherical coordinates)
theta = np.linspace(0, np.pi, 50)
phi = np.linspace(0, 2 * np.pi, 50)
Theta, Phi = np.meshgrid(theta, phi)

# Calculate projections with and without magnetic field
Zx_no_field = fcubicx(Theta, Phi, K0, K1, 0)
Zy_no_field = fcubicy(Theta, Phi, K0, K1, 0)
Zz_no_field = fcubicz(Theta, Phi, K0, K1, 0)

Zx_field = fcubicx(Theta, Phi, K0, K1, 1.5)
Zy_field = fcubicy(Theta, Phi, K0, K1, 1.5)
Zz_field = fcubicz(Theta, Phi, K0, K1, 1.5)

# Create figure for the 3D plots
fig = plt.figure(figsize=(12, 6))
fig.suptitle('Cristal cúbico isotrópico', size=15)

# Plot without magnetic field (H = 0)
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(Zx_no_field, Zy_no_field, Zz_no_field, cmap='copper', edgecolor='k', linewidth=0.5)
ax1.set_title('H = 0 (Campo cero)')
ax1.view_init(elev=30, azim=-120)  # Set view so 
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z')


# Plot with magnetic field (H = 1.5)
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(Zx_field, Zy_field, Zz_field, cmap='copper', edgecolor='k', linewidth=0.5)
ax2.set_title('H = 1.5 (Campo fuerte)')
ax2.view_init(elev=30, azim=-120)  # Set view so 
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('z')

# Show plot
plt.tight_layout()
plt.show()
```
